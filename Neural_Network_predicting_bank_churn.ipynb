```python
import pandas as pd
import numpy as np
import tensorflow as tf
data=pd.read_csv('churn_modelling.csv')
df=pd.DataFrame(data)
print(df)
```

```python
X=df.iloc[:,3:-1].values
y=df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X[:,2]=le.fit_transform(X[:,2])
```

```python
print(X)
```

```python
X[:,1]=le.fit_transform(X[:,1])
```

```python
print(X[:,1])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=6,activation='relu')) #input layer
```

```python
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))#hidden layer1
```

```python
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))#hidden layer2
```

```python
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))#output layer
```

```python
ann.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
print((ann.predict(sc.transform([[600,0,1,40,3,60000,2,1,1,50000]]))))
```

```python
y_pred=ann.predict(X_test)
y_pred=(y_pred>0.2)
print(np.concatenate([y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)], axis=1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm= confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```
